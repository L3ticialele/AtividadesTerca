Faça consultas no DataFrame para exibir:

1. Nome e time dos jogadores que possuem salário acima de R$
20.0000,00

```python
jogadores[(jogadores['salario_do_jogador'] > 200000)][['nome_do_jogador', 'nome_time_jogador']]
```

2. Nome e salário dos jogadores dos times de Minas Gerais

```python
jogadores[(jogadores['nome_estado_jogador'] == 'MG')][['nome_do_jogador', 'salario_do_jogador']]
```

3. Nome e time dos jogadores cujo nome contenha a letra 'u'

```python
jogadores[jogadores.nome_do_jogador.str.contains('u')]
```

4. Nome, salário e time dos jogadores, ordenados pelo salário em
ordem decrescente.

```python
jogadores.sort_values(by='salario_do_jogador', ascending=False)[['nome_do_jogador', 'salario_do_jogador', 'nome_time_jogador']]
```

5. Nome, salário e time dos jogadores, ordenados pelo nome do time
em ordem crescente, e em seguida pelo salário em ordem
decrescente.

```python
#Time do jogador em ordem crescente
jogadores.sort_values(by=['nome_time_jogador', 'salario_do_jogador'], ascending=[True, False])[['nome_do_jogador', 'salario_do_jogador', 'nome_time_jogador']]
```

6. A quantidade de jogadores por time

```python
jogadores['nome_time_jogador'].value_counts()
```

7. A média salarial por time

```python
print(f'A média salárial por time é igual a: {jogadores['salario_do_jogador'].mean()}')
```
